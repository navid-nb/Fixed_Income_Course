# Caps ad floors

```python
def interest_rate_cap_floor(
    maturity: float,
    frequency: int, 
    rate_K: float,
    notional: float,
    rt: float,
    model: str,
    t: float = 0.0,
    first_reset: float = 0.25,
    is_cap: bool = True,
    **params
) -> float:
    """
    Values Caps/Floors by decomposing into ZCB options.
    Logic strictly follows Exercise 3.6 summation from j=2 to 4.
    """
    # 1. Setup payment schedule using precise Reset Dates
    tau = 1/frequency 
    # Use np.arange for exact step sizes to avoid floating point errors
    reset_dates = np.arange(first_reset, maturity, tau)
    t_j = reset_dates + tau
    
    # 2. Translate Rate Strike to Bond Strike (Eq 1.39)
    K_prime = 1 / (1 + rate_K * tau)
    M_prime = notional * (1 + rate_K * tau)
    option_type = 'put' if is_cap else 'call'
    
    total_val = 0.0
    model = model.lower()

    # 3. Summation of ZCB options
    for j in range(len(t_j)):
        # The option for payment t_j expires at the corresponding reset date
        t_expiry = reset_dates[j]
        
        # Safety: only price future options
        if t_expiry <= t + 1e-10:
            continue
            
        if model == 'vasicek':
            price = vasicek_zcb_option(T=t_j[j], T_call=t_expiry, K=K_prime, 
                                       rt=rt, t=t, option_type=option_type, 
                                       kappa=params['kappa'], theta=params['theta'], sigma=params['sigma'])
        elif model in ['cir', 'cox']:
            price = cir_zcb_option(T=t_j[j], T_call=t_expiry, K=K_prime, 
                                   rt=rt, t=t, option_type=option_type, 
                                   kappa=params['kappa'], theta=params['theta'], sigma=params['sigma'])
        elif model == 'hull_white':
            price = hull_white_zcb_option(T=t_j[j], T_call=t_expiry, K=K_prime, 
                                          rt=rt, t=t, option_type=option_type, 
                                          kappa=params['kappa'], sigma=params['sigma'], 
                                          market_zcb_func=params['market_zcb_func'], 
                                          market_fwd_func=params['market_fwd_func'])
            
        if isinstance(price, pd.DataFrame):
            price = price.iloc[0, 0]
            
        total_val += M_prime * price
        
    return float(total_val)



```

```python
# Cap with maturity 5 years, quarterly payments, strike rate 4%, notional $1,000,000 with first settlement at 0.5 years
# Using Vasicek model with kappa=0.1, theta=0.05, sigma=0.01
v_cap = interest_rate_cap_floor(
    maturity=1.0, 
    frequency=4, 
    rate_K=0.06, 
    notional=100, 
    rt=0.05, 
    model='vasicek', 
    is_cap=True,
    t=0.0,
    first_reset=0.5, # Custom first payment date
    kappa=0.15, theta=0.06, sigma=0.05
)
print(f"Vasicek Cap: ${v_cap:,.2f}")
```

```python
# floor with maturity 3 years, semi-annual payments, strike rate 2%, notional $1,000,000 with first settlement at 1 year
# Using CIR model with kappa=0.15, theta=0.05, sigma=0.02
# Parameters from Exercise 3.6
cap_result = interest_rate_cap_floor(
    maturity=1.0, 
    frequency=4, 
    rate_K=0.06, 
    notional=100, 
    rt=0.05, 
    model='cir', 
    first_reset=0.25, 
    kappa=0.15, theta=0.06, sigma=0.05
)

print(f"CIR Cap Value: {cap_result:.6f}") # Result: 0.039258
```

```python
# Hull-White cap with maturity 10 years, quarterly payments, strike rate 3.5%, notional $1,000,000 with first settlement at 0.5 years
# Calibrated to NSS
nss_p = {'a': 0.04, 'b': -0.02, 'c': 0.01, 'd': 0.02, 'tau': 1.0, 'theta': 2.0}
m_zcb = lambda x: calculate_nss(x, **nss_p)['p_NSS'].iloc[0]
m_fwd = lambda x: calculate_nss(x, **nss_p)['f_NSS'].iloc[0]

hw_cap = interest_rate_cap_floor(
    maturity=10.0, 
    frequency=4, 
    rate_K=0.035, 
    notional=1000000, 
    rt=0.03, 
    model='hull_white', 
    is_cap=True,
    market_zcb_func=m_zcb,
    market_fwd_func=m_fwd,
    kappa=0.15, sigma=0.01
)
print(f"Hull-White Cap: ${hw_cap:,.2f}")
```
